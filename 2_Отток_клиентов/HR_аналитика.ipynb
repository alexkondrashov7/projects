```python
def scatter_plot(data,xname,yname):
    rez=[]
    for i in data[xname].unique():
        s=data[data[xname]==i][yname]
        rez.append(s)
    label=data[xname].unique()
    plt.boxplot(rez,labels=label)
    plt.title(f'–î–∏–∞–≥—Ä–∞–º–º–∞ —Ä–∞–∑–º–∞—Ö–∞ –¥–ª—è {xname} –æ—Ç {yname}')
    plt.xlabel(f'{xname}')
    plt.ylabel(f'{yname}')
    plt.show()
        
            
            
```

```python
x=['dept','level','workload','last_year_promo','last_year_violations']
y=['salary','job_satisfaction_rate','employment_years','supervisor_evaluation']
for row in x:
    for sow in y:
        scatter_plot(data_train,row,sow)
        
        
```

–ó–¥–µ—Å—å –º—ã –ø–æ–¥–µ–ª–∏–ª–∏ —Å–æ—Ç—Ä—É–¥–Ω–∏–∫–æ–≤ –Ω–∞ —Ç–µ, –∫–æ–≥–æ –º–æ–∂–Ω–æ —Å—á–∏—Ç–∞—Ç—å —É–¥–æ–≤–ª–µ—Ç–≤–æ—Ä–µ–Ω–Ω—ã–º, –∏ —Ç–µ—Ö –∫–æ–≥–æ –Ω–µ–ª—å–∑—è

```python
y_test=df['job_satisfaction_rate']
X_test=df.drop('job_satisfaction_rate',axis=1)
```

```python
ohe_columns=['dept','last_year_promo','last_year_violations']
num_columns=['employment_years','salary']
ord_columns=['level','workload']
```

```python
ohe_pipe=Pipeline([
    ('simple_imputer_one',SimpleImputer(missing_values=np.nan,strategy='most_frequent')),
('one',OneHotEncoder(drop='first',handle_unknown='ignore',sparse_output=False))
])
ord_pipe=Pipeline([
    ('simple_imputer_ord_before',SimpleImputer(missing_values=np.nan,strategy='most_frequent')),
('ord',OrdinalEncoder(categories=[['junior','midle','sinior'],['low','medium','high']],
                      handle_unknown='use_encoded_value', unknown_value=np.nan)),
    ('simple_imputer_ord_after',SimpleImputer(missing_values=np.nan,strategy='most_frequent'))
])
data_preprocessor=ColumnTransformer([
    ('one',ohe_pipe,ohe_columns),
    ('ord',ord_pipe,ord_columns),
    ('num',StandardScaler(),num_columns)
],
remainder='passthrough')
```

```python
pipe_final=Pipeline([
    ('prepross',data_preprocessor),
('models',DecisionTreeRegressor(random_state=random_state))])
```

```python
param_distributions=[
    { 'models': [LinearRegression()],
     'prepross__num': [StandardScaler(),MinMaxScaler()]
        },
    {
        'models': [DecisionTreeRegressor(random_state=random_state)],
        'models__max_depth': range (1,50),
        'models__max_features': range (1,50),
        'models__max_leaf_nodes': range (100,1000),
        
        'prepross__num': [StandardScaler(),MinMaxScaler()]
    }
]
```

```python
def smape_score(y_true,y_pred):
    if isinstance(y_true, pd.Series):
        y_true = y_true.values
    if isinstance(y_pred, pd.Series):
        y_pred = y_pred.values
    score=0
    for i in range(len(y_true)):
        score=score+(np.abs(y_true[i]-y_pred[i]))/(0.5*(np.abs(y_true[i])+np.abs(y_pred[i])))
    score=score*100/len(y_true)
    return score   
```

```python
smape_score_=make_scorer(smape_score,greater_is_better=False)
```

```python
model1=RandomizedSearchCV(pipe_final,param_distributions,cv=5,scoring=smape_score_,n_iter=100)
```

```python
data1=data_train1.query('quit=="yes"') #—É–≤–æ–ª–∏–≤—à–∏–π—Å—è —Å–æ—Ç—Ä—É–¥–Ω–∏–∫
data2=data_train1.query('quit!="yes"')
label1='—É–≤–æ–ª–∏–≤—à–∏–π—Å—è —Å–æ—Ç—Ä—É–¥–Ω–∏–∫'
label2='–¥–µ–π—Å—Ç–≤—É—é—â–∏–π —Å–æ—Ç—Ä—É–¥–Ω–∏–∫'
```

```python
def hist_plot22(data1,data2,columns,label1,label2):
    for i in columns:
        plt.figure(figsize=(7,7))
        plt.hist(data1[i],density=True,label=label1)
        plt.hist(data2[i],density=True,alpha=0.7,label=label2)
        plt.title(f'–ì–∏—Å—Ç–æ–≥—Ä–∞–º–º–∞ —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏—è {i}')
        plt.xlabel(f'{i}')
        plt.ylabel('–î–æ–ª—è')
        plt.legend()
        plt.show()
        print(f'–î–ª—è {label1}')
        display(data1[i].describe())
        print(f'–î–ª—è {label2}')
        display(data2[i].describe())
```

```python
x=['salary']
hist_plot22(data1,data2,x,label1,label2)
```

```python
def plot_bar22(data1,data2,columns,label1,label2):
    for i in columns:
        res1=data1[i].value_counts(normalize=True).sort_index()
        res2=data2[i].value_counts(normalize=True).sort_index()
        plt.figure(figsize=(7,7))
        plt.bar(res1.index,res1.values,label=label1)
        plt.bar(res2.index,res2.values,label=label2,alpha=0.7)
        #plt.hist(data1[i],density=True,label=label1)
        #plt.hist(data2[i],density=True,alpha=0.7,label=label2)
        plt.title(f'–î–∏–∞–≥—Ä–∞–º–º–∞ —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏—è {i}')
        plt.xlabel(f'{i}')
        plt.ylabel('–î–æ–ª—è')
        plt.legend()
        plt.show()
        print(f'–î–ª—è {label1}')
        display(data1[i].describe())
        print(f'–î–ª—è {label2}')
        display(data2[i].describe())
```

```python
x=['employment_years','supervisor_evaluation']
plot_bar22(data1,data2,x,label1,label2)
```

### –û–±—É—á–µ–Ω–∏–µ –º–æ–¥–µ–ª–µ–π

**–ó–∞–¥–∞—á–∞** –û–±—É—á–∏–º –Ω–µ—Å–∫–æ–ª—å–∫–æ –º–æ–¥–µ–ª–µ–π –¥–ª—è –∑–∞–¥–∞—á–∏ –±–∏–Ω–∞—Ä–Ω–æ–π –∫–ª–∞—Å—Å–∏—Ñ–∏–∫–∞—Ü–∏–∏ (DecisionTreeClassifier,KNeighborsClassifier,LogisticRegression) –∏ –æ–ø—Ä–µ–¥–µ–ª–∏–º –ª—É—á—à—É—é

```python
ohe_columns=['dept','last_year_promo','last_year_violations']
num_columns=['employment_years','salary']
ord_columns=['level','workload']
ohe_pipe=Pipeline([
    ('simple_imputer_one',SimpleImputer(missing_values=np.nan,strategy='most_frequent')),
('one',OneHotEncoder(drop='first',handle_unknown='ignore',sparse_output=False))
])
ord_pipe=Pipeline([
    ('simple_imputer_ord_before',SimpleImputer(missing_values=np.nan,strategy='most_frequent')),
('ord',OrdinalEncoder(categories=[['junior','midle','sinior'],['low','medium','high']],
                      handle_unknown='use_encoded_value', unknown_value=np.nan)),
    ('simple_imputer_ord_after',SimpleImputer(missing_values=np.nan,strategy='most_frequent'))
])
data_preprocessor=ColumnTransformer([
    ('one',ohe_pipe,ohe_columns),
    ('ord',ord_pipe,ord_columns),
    ('num',StandardScaler(),num_columns)
],
remainder='passthrough')
pipe_final=Pipeline([
    ('prepross',data_preprocessor),
('models',LogisticRegression(random_state=random_state))])
param_distributions=[
    { 'models': [LogisticRegression(solver='saga',random_state=random_state,penalty='l1',max_iter=500)],
     'models__C': range (1,5),
     'prepross__num': [StandardScaler(),MinMaxScaler()]
        },
    {
        'models': [DecisionTreeClassifier(random_state=random_state)],
        'models__max_depth': range (10,1000),
        'models__max_features': range (1,25),
        'models__max_leaf_nodes': range (2,100),
        'prepross__num': [StandardScaler(),MinMaxScaler()]
    },
    {
        'models': [KNeighborsClassifier()],
        'models__n_neighbors': range (5,200),
        'prepross__num': [StandardScaler(),MinMaxScaler()]
    }  
]
```

```python
label=LabelEncoder()
```

```python
y_train1=label.fit_transform(y_train1)
```

```python
print(f'0 –∫–æ–¥–∏—Ä—É–µ—Ç—Å—è: {label.classes_[0]}')
print(f'1 –∫–æ–¥–∏—Ä—É–µ—Ç—Å—è: {label.classes_[1]}')
```

–í–∞–∂–Ω–æ –ø–æ–Ω–∏–º–∞—Ç—å, –∫–∞–∫ –∫–æ–¥–∏—Ä—É–µ—Ç—Å—è —Ü–µ–ª–µ–≤–æ–π –ø—Ä–∏–∑–Ω–∞–∫

```python
model2=RandomizedSearchCV(pipe_final,param_distributions,cv=5,scoring='roc_auc',n_iter=100,random_state=random_state)
```

–ë—É–¥–µ–º –¥–ª—è –ø–æ–¥–±–æ—Ä–∞ –æ–ø—Ç–∏–º–∞–ª—å–Ω—ã—Ö –≥–∏–ø–µ—Ä–ø–∞—Ä–∞–º–µ—Ç—Ä–æ–≤ –∏—Å–ø–æ–ª—å–∑–æ–≤–∞—Ç—å –†–∞–Ω–¥–æ–º–∏–∑–∏—Ä–æ–≤–∞–Ω–Ω—ã–π –ø–æ–∏—Å–∫, –ø–æ—Å–∫–æ–ª—å–∫—É –ø–æ –≤—Ä–µ–º–µ–Ω–∏ –æ–Ω –æ–ø—Ä–µ–¥–µ–ª—è–µ—Ç –±—ã—Å—Ç—Ä–µ–µ. –í –∫–∞—á–µ—Å—Ç–≤–µ –º–µ—Ç—Ä–∏–∫–∏ –≤–æ–∑—å–º–µ–º roc_auc, –ø–æ—Å–∫–æ–ª—å–∫—É –æ–Ω–∞ –Ω–µ –∑–∞–≤–∏—Å–∏—Ç –æ—Ç –ø–æ—Ä–æ–≥–∞ –∫–ª–∞—Å—Å–∏—Ñ–∏–∫–∞—Ü–∏–∏ –∏ —É—Å—Ç–æ–π—á–∏–≤–∞ –∫ –¥–æ–º–∏–Ω–∏—Ä–æ–≤–∞–Ω–∏—é –æ–¥–Ω–æ–≥–æ –∫–ª–∞—Å—Å–∞ –Ω–∞–¥ –¥—Ä—É–≥–∏–º
